```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

service = Service(executable_path="chromedriver.exe")
driver = webdriver.Chrome(service=service)

# URL of the page to scrape
url = 'https://fr.shopping.rakuten.com/nav/Tel-PDA_Telephones-mobiles'

# Open the URL
driver.get(url)

# Wait for the page to load completely
time.sleep(10)  # Adjust the sleep time as necessary

# Get the page source and parse it with BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# Close the WebDriver
driver.quit()

# Selector for the product containers
selector = '.mh-16.productList_childrenLayout_d73'
containers = soup.select(selector)

print(containers)
# Check if containers were found
print(f"Found {len(containers)} product containers.")
```

```python

# Extracting data from each container
data = []

for product in containers:
    title_tag = product.find('p', class_='description_styleTitle_KPO lh-title normal mv0 b di')  # Adjust class based on actual HTML
    neuf_price_tag = product.find('span', class_='f20 dib b')
    occasion_price_tag = product.find('span', class_='b dib f20')
    description=product.find('div',class)

    title = title_tag.text.strip() if title_tag else 'N/A'
    neuf_price = neuf_price_tag.text.strip() if neuf_price_tag else 'N/A'
    occasion_price = occasion_price_tag.text.strip() if occasion_price_tag else 'N/A'

    data.append({
        'title': title,
        'neuf_price': neuf_price,
        'occasion_price': occasion_price
    })

# Print extracted data
for item in data:
    print(f"Title: {item['title']}")
    print(f"Neuf Price: {item['neuf_price']}")
    print(f"Occasion Price: {item['occasion_price']}")
    print()
```

```python
import csv
# Define the CSV file path
csv_file_path = 'extracted_data.csv'

# Write data to CSV file
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['title', 'neuf_price', 'occasion_price'])
    writer.writeheader()
    writer.writerows(data)

print(f"Data has been written to {csv_file_path}")
```
